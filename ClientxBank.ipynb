```python
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes

import hashlib
import random
```

```python
print("=== RSA Key Generation ===")
```

```python

alice_key = RSA.generate(2048)

alice_private_key = alice_key
alice_public_key = alice_key.publickey()

print("Alice RSA Keys Generated Successfully")
```

```python

bob_key = RSA.generate(2048)

bob_private_key = bob_key
bob_public_key = bob_key.publickey()

print("Bob RSA Keys Generated Successfully")
```

```python
print("Alice Public Key:")
print(alice_public_key.export_key().decode())

print("\nBob Public Key:")
print(bob_public_key.export_key().decode())
```

```python
print("=== Diffie-Hellman Key Exchange ===")

p = 23
g = 5

print("Prime number (p):", p)
print("Generator (g):", g)
```

```python

a = random.randint(1, 100)
A = pow(g, a, p)

print("Alice private value (a):", a)
print("Alice public value (A):", A)
```

```python

b = random.randint(1, 100)
B = pow(g, b, p)

print("Bob private value (b):", b)
print("Bob public value (B):", B)
```

```python

B_bytes = str(B).encode()
hash_B = SHA256.new(B_bytes)
bob_signature = pkcs1_15.new(bob_private_key).sign(hash_B)

print("Bob signed his Diffie-Hellman value successfully")
```

```python

try:
    pkcs1_15.new(bob_public_key).verify(hash_B, bob_signature)
    print("Bob signature verified successfully")
except:
    print("Verification failed")
```

```python

A_bytes = str(A).encode()
hash_A = SHA256.new(A_bytes)
alice_signature = pkcs1_15.new(alice_private_key).sign(hash_A)

print("Alice signed her Diffie-Hellman value successfully")
```

```python

try:
    pkcs1_15.new(alice_public_key).verify(hash_A, alice_signature)
    print("Alice signature verified successfully")
except:
    print("Verification failed")
```

```python

alice_shared_secret = pow(B, a, p)
bob_shared_secret = pow(A, b, p)

print("Alice Shared Secret:", alice_shared_secret)
print("Bob Shared Secret:", bob_shared_secret)
```

```python
print("=== Key Derivation ===")


shared_secret_bytes = str(alice_shared_secret).encode()
aes_key = hashlib.sha256(shared_secret_bytes).digest()

print("AES Key Generated Successfully")
print("AES Key Length:", len(aes_key))
```

```python
transaction_data = """
Transaction ID: TXN1001
From: Alice
To: Bob
Amount: 5000 EGP
Status: Approved
"""

with open("transactions.txt", "w") as file:
    file.write(transaction_data)

print("transactions.txt created successfully")
```

```python
with open("transactions.txt", "r") as file:
    plaintext = file.read()

print("Original File Content:")
print(plaintext)
```

```python
print("=== AES Encryption ===")

cipher = AES.new(aes_key, AES.MODE_EAX)
ciphertext, tag = cipher.encrypt_and_digest(plaintext.encode())
nonce = cipher.nonce

print("File encrypted successfully")
```

```python
with open("encrypted.bin", "wb") as file:
    file.write(nonce)
    file.write(tag)
    file.write(ciphertext)

print("Encrypted file saved as encrypted.bin")
```

```python

cipher_hash = SHA256.new(ciphertext)
cipher_signature = pkcs1_15.new(alice_private_key).sign(cipher_hash)

print("Encrypted file signed successfully")
```

```python
try:
    pkcs1_15.new(alice_public_key).verify(cipher_hash, cipher_signature)
    print("Ciphertext signature verified successfully")
except:
    print("Signature verification failed")
```

```python
print("=== AES Decryption ===")

decrypt_cipher = AES.new(aes_key, AES.MODE_EAX, nonce=nonce)
decrypted_data = decrypt_cipher.decrypt_and_verify(ciphertext, tag)

print("File decrypted successfully")
```

```python
with open("decrypted.txt", "wb") as file:
    file.write(decrypted_data)

print("Decrypted file saved successfully")

print("\nDecrypted Content:\n")
print(decrypted_data.decode())
```

```python
if plaintext == decrypted_data.decode():
    print("SUCCESS: Original and decrypted files match")
else:
    print("ERROR: Files do not match")
```
